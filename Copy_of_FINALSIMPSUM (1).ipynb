```python
!pip install PyPDF2

import PyPDF2

def pdf_to_text(pdf_file_path, text_file_path):
    try:
        Open the PDF file
        with open(pdf_file_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)

             #Initialize an empty string to store the text
            text = ""

            # Loop through each page of the PDF
            for page_num in range(len(pdf_reader.pages)):
                page = pdf_reader.pages[page_num]
                text += page.extract_text()

            # Write the extracted text to a text file
            with open(text_file_path, 'w', encoding='utf-8') as text_file:
                text_file.write(text)

            print(f"PDF text extracted and saved to {text_file_path}")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

if __name__ == "__main__":
    pdf_file_path = "MANU_DE_0232_2022.pdf"  # Replace with the path to your PDF file
    text_file_path = "output.txt"  # Replace with the desired output text file path
    pdf_to_text(pdf_file_path, text_file_path)
```

```python
text_file_path = '/content/manu.txt'

# Open the text file
with open(text_file_path, 'r', encoding='utf-8') as text_file:
    text = text_file.read()
```

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(
        separator="\n",
        chunk_size=1000,
        chunk_overlap=100,
        length_function=len
    )
chunks = text_splitter.split_text(text)
```

```python
#for simp
input_text = chunks[0]

# Remove spaces from the input text
input_text_no_spaces = input_text.replace(" ", "")

# Split the text into paragraphs
paragraphs = input_text.split('\n')

# Initialize a variable to store the extracted text
extracted_text = ""

# Iterate through paragraphs until the word "judgment" (space-insensitive) is found
for paragraph in paragraphs:
    extracted_text += paragraph + '\n'
    if "JUDGMENT" in paragraph.upper().replace(" ", ""):
        break

response_text=""
response_text=extracted_text
```

```python
##for simp

for part in chunks:
    prompt_template = f'''SYSTEM: You are a helpful, respectful, and legal assistant.
     Always answer as helpfully. The paragraphs are too complex, making it difficult for the reader to understand the main points.
     Can you simplify them keeping all the key points intact. Make meaningful statements.
     Strictly follow the below instructions all the time.
     First line should strictly contain only " Here is the simplified content : " , and from the next line strictly generate the simplified content.

    USER: {part}
    ASSISTANT:
    '''
    response = lcpp_llm(
        prompt=prompt_template,
        max_tokens=4096,
        temperature=0.9,
        top_p=0.95,
        repeat_penalty=1.5,
        top_k=50,
        stop=['USER:'],  # Dynamic stopping when such a token is detected.
        echo=False  # Return the prompt
    )
    # Extract the generated response text without unwanted messages
    response_text += response["choices"][0]["text"]


```

```python
#for simp
import re
import textwrap

input_text = response_text

text_lines = input_text.split('\n')

modified_content = []

patterns = [ #insert any lines which you'd like to avoid in your output

]

pattern_regexes = [re.compile(r'\s*' + re.escape(pattern) + r'\s*', re.IGNORECASE) for pattern in patterns]


for line in text_lines:
    modified_line = line
    for pattern_regex in pattern_regexes:
        modified_line = pattern_regex.sub("", modified_line)
    modified_content.append(modified_line)


final_text = '\n'.join(modified_content)

line_width = 80
wrapped_text = textwrap.fill(final_text, width=line_width)

print(wrapped_text)

text_to_save = wrapped_text

file_path = 'output1.txt'

with open(file_path, 'w', encoding='utf-8') as file:
    file.write(text_to_save)

print(f'Text has been saved to {file_path}')
```
